```python
#데이터 파일들을 하나로 병함

excel_files = ['6-7 ~ 6-12.xlsx', '6-13 ~ 10-5.xlsx', '10-6 ~ 10-13.xlsx', '10-14 ~ 11-1.xlsx', '11-2 ~ 2-4.xlsx', '2-5 ~ 5-9.xlsx', '5-10 ~ 6-7.xlsx']

all_data = []  # 모든 데이터를 저장할 리스트

for excel_file in excel_files:
    data = pd.read_excel(excel_file, sheet_name=None, header=None)
    all_data.append(data)  # 데이터를 리스트에 추가

all_data
```

```python
#데이터에 맞는 자료형으로 변환 후 이상치 처리

import numpy as np

def remove_outliers(data):
    # 문자열 숫자로 변환 후 ',' 제거
    data[0] = data[0].str.replace('₩', '').str.replace(',', '').astype(float)
    
    # 이상치 제외를 위한 통계 값 계산
    Q1 = np.percentile(data[0], 25)  # 1사분위수
    Q3 = np.percentile(data[0], 75)  # 3사분위수
    IQR = Q3 - Q1  # IQR 계산
    lower_bound = Q1 - 1.5 * IQR  # 이상치 제외를 위한 하한값
    upper_bound = Q3 + 1.5 * IQR  # 이상치 제외를 위한 상한값
    
    # 이상치 제외된 데이터 추출
    filtered_data = data[(data[0] >= lower_bound) & (data[0] <= upper_bound)]
    
    return filtered_data

filtered_all_data = []  # 이상치가 제외된 데이터를 저장할 리스트

for data_dict in all_data:
    filtered_data_dict = {}  # 이상치가 제외된 데이터를 저장할 딕셔너리
    for sheet_name, sheet_data in data_dict.items():
        # 이상치 제외된 데이터 생성
        filtered_data = remove_outliers(sheet_data)
        filtered_data_dict[sheet_name] = filtered_data  # 이상치가 제외된 데이터를 딕셔너리에 저장
    filtered_all_data.append(filtered_data_dict)  # 딕셔너리를 리스트에 추가

filtered_all_data
```

```python
#날짜별 평균 가격

averages = []  # 각 날짜별 평균 가격을 저장할 리스트

for data_dict in filtered_all_data:
    for sheet_name, sheet_data in data_dict.items():
        # 첫 번째 열을 가격 정보로 사용하여 평균 계산
        average_price = sheet_data[0].mean()
        averages.append((sheet_name, average_price))  # (날짜, 평균 가격) 튜플을 리스트에 추가

averages
```

```python
#계절 부여

seasons = {
    1: '겨울',
    2: '겨울',
    3: '봄',
    4: '봄',
    5: '봄',
    6: '여름',
    7: '여름',
    8: '여름',
    9: '가을',
    10: '가을',
    11: '가을',
    12: '겨울'
}

season_data = []  # 계절별 가격 데이터를 저장할 리스트

for date, average_price in averages:
    month = int(date.split()[0].split('월')[0])  # 날짜에서 월 정보 추출
    season = seasons[month]  # 계절 정보 가져오기
    season_data.append((season, average_price))  # (계절, 평균 가격) 튜플을 리스트에 추가

season_data
```

```python
import pandas as pd

# season_data를 DataFrame으로 변환
df = pd.DataFrame(season_data, columns=['Season', 'Value'])

# 계절로 그룹화
grouped_data = df.groupby('Season')

# 그룹화된 데이터 확인
for season, group in grouped_data:
    print(season)
    print(group)
    print()
```

```python
#일원분산분석 시행

import pandas as pd
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# 계절별 가격 데이터를 저장한 리스트를 DataFrame으로 변환
df = pd.DataFrame(season_data, columns=['Season', 'Value'])

# 일원분산분석 수행
fvalue, pvalue = f_oneway(df[df['Season'] == '봄']['Value'],
                          df[df['Season'] == '여름']['Value'],
                          df[df['Season'] == '가을']['Value'],
                          df[df['Season'] == '겨울']['Value'])

# 결과 출력
print("One-way ANOVA")
print("F-value:", fvalue)
print("p-value:", pvalue)

# 사후분석 수행 (Tukey's HSD test)
tukey_result = pairwise_tukeyhsd(df['Value'], df['Season'])
print("Post-hoc (Tukey's HSD)")
print(tukey_result)
```

```python
One-way ANOVA 결과:

F-value: 34.95432277938549
p-value: 7.221314517683334e-20
One-way ANOVA는 각 계절별 가격 데이터를 사용하여 그룹 간의 평균 차이가 있는지를 검정하는 통계 분석 방법입니다. F-value는 그룹 간의 분산 비율을 나타내며, p-value는 귀무가설(그룹 간 평균이 동일하다는 가설)이 유의 수준에서 기각되는지를 나타냅니다. 이 경우, p-value가 매우 작은 값(7.221314517683334e-20)으로 나타났으므로 귀무가설을 기각할 수 있습니다. 즉, 적어도 하나의 계절 그룹 간에는 평균 가격 차이가 존재한다는 결론을 내릴 수 있습니다.

사후분석 (Tukey's HSD test) 결과:

가을과 겨울: 가격 차이가 유의미함 (reject=True)
가을과 봄: 가격 차이가 유의미함 (reject=True)
가을과 여름: 가격 차이가 유의미하지 않음 (reject=False)
겨울과 봄: 가격 차이가 유의미하지 않음 (reject=False)
겨울과 여름: 가격 차이가 유의미함 (reject=True)
봄과 여름: 가격 차이가 유의미함 (reject=True)
사후분석은 One-way ANOVA에서 그룹 간의 평균 차이가 있는 것으로 나타난 경우, 어떤 그룹들 간에 유의미한 차이가 있는지를 검정하는 방법입니다. 이 결과를 보면, 가을과 겨울, 가을과 봄, 겨울과 여름, 그리고 봄과 여름 간의 가격 차이는 통계적으로 유의미하다는 것을 알 수 있습니다. 따라서, 이 네 개의 계절 간에는 평균 가격이 서로 다르다고 할 수 있습니다.

그러나 가을과 여름, 겨울과 봄 간의 가격 차이는 유의미하지 않다는 결과가 나타났습니다. 이는 해당 계절 간의 평균 가격이 통계적으로 유의미한 차이를 보이지 않는다는 것을 의미합니다.

따라서, 이 분석 결과는 각 계절별 가격 차이에 대해 유의미한 결과를 제시하며, 특정 계절 간의 가격 차이가 다른 계절들과 비교했을 때 유의미하다는 것을 나타냅니다. 이를 통해 계절에 따른 항공권 가격의 변동을 파악하고, 가격 예측이나 계절별 마케팅 전략 수립에 도움을 줄 수 있습니다.
```

```python
박스 (Box):

상자는 데이터의 중앙값(median)을 나타냅니다. 상자 안의 가로선은 중앙값을 표시합니다.
상자의 상단은 데이터의 75번째 백분위수 (3분위수, 75th percentile)를 나타내며, 하단은 데이터의 25번째 백분위수 (1분위수, 25th percentile)를 나타냅니다.
상자의 높이는 데이터의 중간 50% 범위인 사분위 범위 (interquartile range, IQR)를 나타냅니다.
수염 (Whiskers):

수염은 데이터의 전체 범위를 표현합니다. 일반적으로 1.5 * IQR 범위까지를 수염으로 표시합니다.
이상치 (outlier)로 간주되는 데이터는 수염 밖의 개별적인 점으로 표시될 수 있습니다.
이상치 (Outliers):

박스플롯에서 이상치는 일반적인 데이터 분포에서 벗어난 극단적인 값으로 간주됩니다.
이상치는 박스와 수염 영역을 벗어나 개별적인 점으로 표시됩니다.
위의 박스플롯에서 해석할 수 있는 내용은 다음과 같습니다:

겨울(Winter)과 봄(Spring)의 가격 분포는 유사한 형태를 가지고 있습니다. 중앙값은 봄이 겨울보다 높지만 큰 차이는 없습니다.
여름(Summer)과 가을(Autumn)의 가격 분포는 유사한 형태를 가지고 있습니다. 중앙값은 여름이 가을보다 약간 높습니다.
봄과 여름, 가을과 겨울 사이에는 가격 차이가 있는 이상치(outlier)가 존재합니다. 이러한 이상치는 해당 계절의 일부 특이한 가격 상황을 나타낼 수 있습니다.
박스플롯을 통해 각 계절별로 가격의 분포와 차이를 시각적으로 파악할 수 있습니다. 이를 통해 계절별 항공권 가격의 특징과 이상치를 확인할 수 있습니다.
```

```python
import re

weekday_data = []  # 요일별 가격 데이터를 저장할 리스트

for date, average_price in averages:
    weekday = re.findall('\((.*?)\)', date)[0]  # 괄호 안에 있는 요일 추출
    weekday_data.append((weekday, average_price))  # (요일, 평균 가격) 튜플을 리스트에 추가

weekday_data
```

```python
#주말과 평일간의 차이


from scipy import stats


weekday_prices = []  # 평일 가격을 저장할 리스트
weekend_prices = []  # 주말 가격을 저장할 리스트

for date, price in weekday_data:
    if '토' in date or '일' in date:  # 주말인 경우
        weekend_prices.append(price)
    else:  # 평일인 경우
        weekday_prices.append(price)

# 평일과 주말의 가격 차이 분석
mean_weekday = pd.Series(weekday_prices).mean()
mean_weekend = pd.Series(weekend_prices).mean()
t_statistic, p_value = stats.ttest_ind(weekday_prices, weekend_prices)

# 결과 출력
print("평일 평균 가격:", mean_weekday)
print("주말 평균 가격:", mean_weekend)
print("평일과 주말 가격 차이의 t-statistic:", t_statistic)
print("평일과 주말 가격 차이의 p-value:", p_value)
```

```python
평일 평균 가격: 2,220,389.3929140638
주말 평균 가격: 2,284,935.6273025377
t-statistic: -2.226552100469077
p-value: 0.026586954043756165
주어진 결과를 해석해보면 다음과 같습니다:

평일과 주말의 평균 가격 차이: 주말의 평균 가격이 평일의 평균 가격보다 약간 높게 나타났습니다. (주말 평균 가격이 64,546.2343884739 원 더 높음)

t-statistic: -2.226552100469077

t-statistic은 두 집단의 평균 간의 차이를 나타내는 통계값입니다.
여기서 t-statistic 값이 음수인 경우, 주말의 평균 가격이 평일의 평균 가격보다 낮은 것을 의미합니다.
t-statistic 값이 크면 클수록 두 집단 간의 차이가 유의미하다고 해석할 수 있습니다.
p-value: 0.026586954043756165

p-value는 주어진 데이터에서 관찰된 차이가 우연에 의한 것인지 아니면 실제 차이가 있는 것인지를 판단하는데 사용됩니다.
여기서 p-value는 0.026586954043756165로 나타났는데, 이는 주말과 평일의 가격 차이가 우연에 의한 것이 아니라는 것을 의미합니다.
일반적으로 0.05보다 작은 p-value는 유의미한 차이를 나타내는 것으로 간주됩니다.
따라서, 주말과 평일 간의 가격 차이는 통계적으로 유의미하다고 볼 수 있습니다. 하지만, 결과의 해석은 분석하고자 하는 도메인의 특성과 문맥에 따라 달라질 수 있으므로, 해당 결과를 도메인 전문가와 함께 검토하고 심층적인 분석을 수행하는 것이 좋습니다.
```

```python
#날짜에 관계없이 데이터 병합

import pandas as pd

excel_files = ['6-7 ~ 6-12.xlsx', '6-13 ~ 10-5.xlsx', '10-6 ~ 10-13.xlsx', '10-14 ~ 11-1.xlsx', '11-2 ~ 2-4.xlsx', '2-5 ~ 5-9.xlsx', '5-10 ~ 6-7.xlsx']

all_data = []

for excel_file in excel_files:
    data = pd.read_excel(excel_file, sheet_name=None, header=None)

    # 각 시트의 데이터를 하나의 데이터프레임으로 합치기
    combined_data = pd.concat(data.values())

    all_data.append(combined_data)

# 모든 데이터를 하나의 데이터프레임으로 합치기
merged_data = pd.concat(all_data)

# 가격 열 변환: 숫자로 변환하고 통화 기호와 쉼표 제거
merged_data[0] = merged_data[0].str.replace('₩', '').str.replace(',', '').astype(float)


# 결과 출력
print(merged_data)
```

```python
#출발시간 24시 형식으로 변환

# 시간을 표현하는 열 선택 (예: 1열)
time_column = merged_data[1]

# 오전/오후 정보와 시간 정보 분리
am_pm = time_column.str.split().str[0]  # 오전/오후 정보 추출
time_info = time_column.str.split().str[1]  # 시간 정보 추출

# 시간 정보를 24시간 형식으로 변환
hour_minute = time_info.str.split(':')
hours = hour_minute.str[0].astype(int)
minutes = hour_minute.str[1].astype(int)
hours = hours.mask(am_pm == '오후', hours + 12)  # 오후인 경우 12를 더해 24시간 형식으로 변환
hours = hours.mask(am_pm == '오전', hours % 12)  # 오전인 경우 12로 나눠 24시간 형식으로 변환

# 변환된 시간 정보로 열 업데이트
merged_data[1] = hours.astype(str).str.zfill(2) + ':' + minutes.astype(str).str.zfill(2)

# 결과 출력
print(merged_data)
```

```python
#도착시간 24시 형식으로 변환

# 시간을 표현하는 열 선택 (예: 2열)
time_column_2 = merged_data[2]

# 오전/오후 정보와 시간 정보 분리
am_pm_2 = time_column_2.str.split().str[0]  # 오전/오후 정보 추출
time_info_2 = time_column_2.str.split().str[1]  # 시간 정보 추출

# 시간 정보를 24시간 형식으로 변환
hour_minute_2 = time_info_2.str.split(':')
hours_2 = hour_minute_2.str[0].astype(int)
minutes_2 = hour_minute_2.str[1].astype(int)
hours_2 = hours_2.mask(am_pm_2 == '오후', hours_2 + 12)  # 오후인 경우 12를 더해 24시간 형식으로 변환
hours_2 = hours_2.mask(am_pm_2 == '오전', hours_2 % 12)  # 오전인 경우 12로 나눠 24시간 형식으로 변환

# 변환된 시간 정보로 열 업데이트
merged_data[2] = hours_2.astype(str).str.zfill(2) + ':' + minutes_2.astype(str).str.zfill(2)

# 결과 출력
print(merged_data)
```

```python
#시간대별 그룹화

# 새로운 열을 추가하여 시간대 정보 기록
merged_data['Time_Group'] = ''

# 시간대 그룹화 함수 정의
def group_by_time(row):
    hour = int(row[1].split(':')[0])  # 출발 시간의 시간 정보 추출
    if 0 <= hour < 6:  # 0:00 - 6:00: 새벽
        return '새벽'
    elif 6 <= hour < 12:  # 6:00 - 12:00: 오전
        return '오전'
    elif 12 <= hour < 18:  # 12:00 - 18:00: 오후
        return '오후'
    else:  # 18:00 - 0:00: 밤
        return '밤'

# 시간대 그룹화 수행
merged_data['Time_Group'] = merged_data.apply(group_by_time, axis=1)

# 결과 출력
print(merged_data)
```

```python
import numpy as np
import scipy.stats as stats

# 시간대별 가격 평균 계산
time_group_means = merged_data.groupby('Time_Group')[0].mean()

# 시간대별 가격 차이 분석
anova_result = stats.f_oneway(
    merged_data.loc[merged_data['Time_Group'] == '새벽', 0],
    merged_data.loc[merged_data['Time_Group'] == '오전', 0],
    merged_data.loc[merged_data['Time_Group'] == '오후', 0],
    merged_data.loc[merged_data['Time_Group'] == '밤', 0]
)

# 결과 출력
print("시간대별 평균 가격:")
print(time_group_means)
print("\n시간대별 가격 차이 분석 (One-way ANOVA):")
print("F-value:", anova_result.statistic)
print("p-value:", anova_result.pvalue)
```

```python
분석 결과를 살펴보면, 새벽 시간대의 평균 가격이 가장 높고(4,663,765), 밤 시간대의 평균 가격이 가장 낮은 것(2,483,892)으로 나타났습니다. 오전 시간대와 오후 시간대의 평균 가격은 비슷한 수준으로 나타났습니다.

One-way ANOVA 결과에 따르면, 시간대별 가격 차이에 대한 F-value는 446.68이며, p-value는 3.43e-287입니다. 이는 유의수준 0.05에서 매우 작은 p-value를 가지므로, 시간대별 가격 간에 통계적으로 유의한 차이가 있다고 할 수 있습니다. 다시 말해, 시간대에 따라 가격이 다르게 형성되었으며, 이 차이는 우연히 발생한 것이 아니라 의미 있는 차이라고 할 수 있습니다.

따라서, 분석 결과에 따르면 시간대별로 가격 차이가 존재하며, 특히 새벽 시간대의 가격이 다른 시간대보다 비교적 높은 것으로 나타났습니다.
```

```python
import pandas as pd
from scipy.stats import f_oneway

# 도착시간대와 출발시간대를 기준으로 그룹화하여 가격의 평균 계산
grouped_data = merged_data.groupby(['Time_Group_Arrival', 'Time_Group'])[0].mean()

# 도착시간대와 출발시간대 조합의 개수 계산
num_combinations = len(grouped_data)

# 도착시간대와 출발시간대의 조합 및 해당 가격 평균 출력
print("도착시간대와 출발시간대 조합별 가격 평균:")
print(grouped_data)

# 도착시간대와 출발시간대의 조합별 데이터 수집
data_by_combinations = []
for combination in grouped_data.index:
    data = merged_data[(merged_data['Time_Group_Arrival'] == combination[0]) & (merged_data['Time_Group'] == combination[1])][0]
    data_by_combinations.append(data)

# 가격 차이 분석 (도착시간대와 출발시간대 조합, One-way ANOVA)
f_value, p_value = f_oneway(*data_by_combinations)

# 결과 출력
print("\n도착시간대와 출발시간대 조합별 가격 차이 분석 (One-way ANOVA):")
print(f"F-value: {f_value}")
print(f"p-value: {p_value}")
```

```python
#시트에 날짜 정보 추가

import pandas as pd

excel_files = ['6-7 ~ 6-12.xlsx', '6-13 ~ 10-5.xlsx', '10-6 ~ 10-13.xlsx', '10-14 ~ 11-1.xlsx', '11-2 ~ 2-4.xlsx', '2-5 ~ 5-9.xlsx', '5-10 ~ 6-7.xlsx']

all_data = []

for excel_file in excel_files:
    data = pd.read_excel(excel_file, sheet_name=None, header=None)

    # 각 시트의 데이터를 하나의 데이터프레임으로 합치기
    combined_data = pd.concat(data.values())

    # 시트 이름을 저장하는 열 추가
    combined_data['Sheet Name'] = [sheet_name for sheet_name in data.keys() for _ in range(len(data[sheet_name]))]

    all_data.append(combined_data)

# 모든 데이터를 하나의 데이터프레임으로 합치기
merged_data = pd.concat(all_data)

# 가격 열 변환: 숫자로 변환하고 통화 기호와 쉼표 제거
merged_data[0] = merged_data[0].str.replace('₩', '').str.replace(',', '').astype(float)

# 결과 출력
print(merged_data)
```

```python
merged_data['Month'] = merged_data['Sheet Name'].str.extract('(\d+)월')
merged_data['Day'] = merged_data['Sheet Name'].str.extract('(\d+)일')

# Month 열 변환
merged_data['Month'] = merged_data['Month'].str.replace('월', '').astype(int)

# Day 열 변환
merged_data['Day'] = merged_data['Day'].str.replace('일', '').astype(int)



print(merged_data)
```
